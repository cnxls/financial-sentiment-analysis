# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

#### 1. Load Data

```python
df = pd.read_csv("../data/raw/phrasebank_raw.csv")
print(df)
```

```python
df.info()
```

#### 2. Class Distribution

```python
counts = df["sentiment"].value_counts()
print(counts)
```

```python
plt.bar(counts.index,counts.values)
plt.title('Sentiment class distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

#### 3. Missing Values & Duplicates

```python
missing = df.isnull().sum()
missing_columns = missing[missing > 0]
print(missing_columns)
```

```python
len(df['sentence'].unique()) < len(df.index)
```

```python
df.drop_duplicates(subset=['sentence'], inplace=True)
```

```python
df.info()
```

#### 4. Word Count Distribution

```python
df["word_count"] = df["sentence"].apply(lambda x: len(x.split()))
plt.hist(df["word_count"], bins=50)
plt.title('Wordcount distribution')
plt.xlabel('N of words')
plt.ylabel('Count')
plt.show()
```

#### 5. Character Count Distribution

```python
df["char_count"] = df["sentence"].apply(lambda x: len(x))
plt.hist(df["char_count"], bins=50)
plt.title('Character count distribution')
plt.xlabel('N of chars')
plt.ylabel('Count')
plt.show()
```

#### 6. Top Bigrams by Sentiment

```python
clean_df = pd.read_csv('../data/processed/clean_df.csv')
positives = clean_df[clean_df["sentiment"] == "positive"]
negatives = clean_df[clean_df["sentiment"] == "negative"]
neutrals = clean_df[clean_df["sentiment"] == "neutral"]
```

```python
def remove_placeholders(text):
      return re.sub(r'<[^>]+>', '', text).strip()
```

```python
positive_words = " ".join(positives["clean_text"])
positive_words = remove_placeholders(positive_words).split()
```

```python
from nltk.util import ngrams
from collections import Counter
positive_bigrams = Counter(ngrams(positive_words, 2))
```

```python
wrds = ['-'.join(x) for x, c in positive_bigrams.most_common(30)]
wdth = [c for x, c in positive_bigrams.most_common(30)]

plt.barh(wrds, wdth, color='blue')
```
