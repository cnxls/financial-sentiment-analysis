```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/phrasebank_raw.csv")
print(df)
```

```python
df.info()
```

```python
counts = df["sentiment"].value_counts()
print(counts)
```

```python
plt.bar(counts.index,counts.values)
plt.title('Sentiment class distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
missing = df.isnull().sum()
missing_columns = missing[missing > 0]
print(missing_columns)
```

```python
len(df['sentence'].unique()) < len(df.index)
```

```python
df.drop_duplicates(subset=['sentence'], inplace=True)
```

```python
df.info()
```

Wordcount to sentiment

```python
df["word_count"] = df["sentence"].apply(lambda x: len(x.split()))
plt.hist(df["word_count"], bins=50)
plt.title('Wordcount distribution')
plt.xlabel('N of words')
plt.ylabel('Count')
plt.show()
```

Charcount to sentiment

```python
df["char_count"] = df["sentence"].apply(lambda x: len(x))
plt.hist(df["char_count"], bins=50)
plt.title('Character count distribution')
plt.xlabel('N of chars')
plt.ylabel('Count')
plt.show()
```
