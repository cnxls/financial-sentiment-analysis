```python
import pandas as pd
```

```python
with open("../data/raw/Sentences_50Agree.txt" ,"r", encoding="latin-1") as f:
    lines = f.readlines()
    rows = []
    for line in lines:
        line = line.strip()
        last_at = line.rfind('@')
        sentence = line[:last_at]
        label = line[last_at+1:]
        row = {
            "sentence" : sentence,
            "sentiment": label,
        }
        rows.append(row)

    print(rows)
```

```python
df = pd.DataFrame(rows)
print(df)
```

```python
df.to_csv("../data/raw/phrasebank_raw.csv", index = False)
```
